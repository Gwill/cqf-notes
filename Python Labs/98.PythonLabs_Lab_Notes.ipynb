<h3 id="Rank-1-Array">Rank 1 Array<a class="anchor-link" href="#Rank-1-Array">¶</a></h3>



<ul>
<li>This is a non-intutive structure</li>
<li>Does not behave consistently</li>
<li>It is neither a row vector nor a column vector</li>
<li>For programming exercises, do not use Rank 1 array. </li>
<li>Always define a row (1,n) or column(n,1) vector</li>
</ul>


```

import numpy as np
from numpy import linalg as LA

```

```

a = np.random.randn(5)
a

```

```

a.shape

```

```

a.T

```

```

np.dot(a,a.T) # the outer product is not matrix

```

```

b = np.random.randn(5,1) # this is a column vector
b

```

```

b.shape

```

```

np.dot(b,b.T) # outer product is a matrix

```

```

c = np.random.randn(1,5) # this is a row vector
c

```

```

np.dot(c.T, c) # outer product is a matrix

```


<h3 id="Reshape">Reshape<a class="anchor-link" href="#Reshape">¶</a></h3>


```

w = np.arange(5)
w

```

```

w.shape

```

```

w1 = w.reshape((5,1))
w1

```

```

w1.shape # column vector

```

```

w2 = w.reshape([-1,1])
w2

```

```

w2.shape

```

```

w3 = np.reshape(w,[-1,1])
w3

```

```

w3.shape

```


<ul>
<li>NumPy newaxis is used to increase the dimension of the existing array. </li>
<li>Thus, 1D array will become 2D and 2D array will become 3D, ...</li>
</ul>


```

w4 = w[:,np.newaxis] # used to increase the dimension: n+1
w4 

```

```

w4.shape

```

```

w4.flatten() # reducing the dimension

```
